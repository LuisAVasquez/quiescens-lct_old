# Structure of the Semester

> **There was no gap between the end of the first semester and the start of the second one.**

Just after we finished our exams for the first semester, we started the second one. For those of us who were already burning out, this was very inconvenient. Honestly, I was not in a condition to study during the first week of the second semester... 

Luckily for us, we had already asked the Coordinator of the local Master (not the LCT Coordinator!) through the student representatives to stop having the [*Alternance*](https://luisavasquez.github.io/quiescens-lct/lct%20year%201/2021/12/07/first_semester.html#Alternance) system, in order to have a lighter workload. Since none of us where working in a company or a lab under the *Alternance* system, the Coordinator agreed.



The courses had the same overall characteristics as [the ones in the first semester](https://luisavasquez.github.io/quiescens-lct/lct%20year%201/2021/12/07/first_semester.html#Courses), which I reproduce here:


- the professor has total decision power over the content *and* organisation of the course, even if there is a published course description. This may contrast with other academic systems (like the UK, USA, Germany...)
- don't expect to get a syllabus
- classrooms may change every couple of weeks
- expect a final project, an exam, or both
- some courses are shared with students from the [Cognitive Sciences Master](https://idmc.univ-lorraine.fr/courses/master-1-sciences-cognitives/) or from the [Sciences du Langage Master](https://formations.univ-lorraine.fr/fr/master/1736-master-analyse-et-traitement-du-lexique.html)
- the Cognitive Sciences students are very loud
- there will always be some deadline coming
- most courses don't go deep into the material, instead, they focus on projects and homework: Even if the material is not hard, it is the workload what makes it difficult
- some courses are at the *Campus Lettres et Sciences Humaines* (Humanities Campus), three blocks away from the IDMC


> Something that was different from this previous semester was that **this semester has several more homework and group projects** than the first one.

There were no electives, all modules were mandatory.

# Courses

## Calculability and Complexity

**Where?**
- IDMC

**Grading**
- 1 exam | 100%
- Several worksheets that may or may not add to your final grade (this was never clear).

**Description**

This is a Pure Mathematics course introducing the analysis of the efficiency algorithms. It covers a formal definition of algorithm, purposes of algorithms, famous algorithms, and limitations of algorithms.

Every two weeks there is a lecture, and between two lectures there is a tutorial session. After each lecture, a new worksheet is available. You have one week to turn it in. The worksheet will be solved in the tutorial session.

**Topics**
- Introduction to set theory
- Turing Machines
- Composition of Turing Machines
- The Halting Problem
- Algorithmic complexity
- Recursion

**Comment**

> **A very difficult course**: Most of the time most of our group didn't know what was going on on the whiteboard.

Rigorous proofs and mathematical reasoning are a big part of this course. Don't make the mistake of working alone on the worksheets: They are very long!

The conceptual part of the course (formal algorithsm, Turing Machines, Halting problem) are tough to grasp for everyone. Luckily, there are tons of materials and tutorials available online. Sometimes the professor's slides or explanations were confusing for those without experience in the subject.

Something required for the course but not covered by it was the *computation of limits*. If you've taken an Introduction to  Calculus course, you should be fine, otherwise, start looking for tutorials online.

Don't be afraid to ask your classmates for help, this is a very tough course!

## Formal Languages

**Where?**
- IDMC

**Grading**
- 1 Exam | 100%

**Description**
This is a Pure Mathematics course introducing Finite State Automata, Regular Expressions, and Formal Grammars in a rigorous way, and showing several equivalences and transformations among them.

The concepts from this course are helpful for those interested in rule-based approaches to NLP.


**Topics**
- Regular languages, Phrase Structure Grammars, Chomsky hierarchy
- Deterministic and non-Deterministic Finite State Automata
- Automata with empty transitions
- Regular expressions, equivalence with type-3 grammars and FSAs
- Pumping lemma for regular languages
- Pushdown automata and equivalence to Context-Free grammars
- Parsing with the CKY and Earley algorithms

**Comment**

> **A very difficult course**: Most of the time most of our group didn't know what was going on on the whiteboard.

Like the *Formal Logic* course from the first semester, this course mainly consists of rigurous definitions and proofs. Again, those students who had had experience with it had to coach everyone else for this course. 

We had already studied most of these systems in the *Methods for NLP* course in the first semester, but only in an intuitive way.

The most important part is to learn to convert from one system to another, say, CFG to push-down automata. Get a hang of it by doing exercises.

The exam consists of exercises that are similar to the ones seen in class.

## Lexicology

**Where?**
- IDMC

**Grading**
- 1 test in class | 30%
- Project | 70%

**Description**

Introduction to lexicology using the Explanatory Combinatorial Linguistics approach, for English.

**Topics**

- Semiotics
- Linguistic sign
- Lexical unit
- Phraseology
- Semantic actants
- Polysemy
- Structure of the lexicon
- Describing lexical meaning
- Government patterns
- Introduction to Lexical functions
- Paradigmatic lexical functions
- Syntagmatic lexical functions
- Lexicographic exemplifications
- Lexical systems

**Comment**

> This course is very heavy in theoretical linguistics!

This course has an associated course called "Syntactical structures".

IMO, the professor was the most organized one in the masters: he actually followed a syllabus and gave readings.

This course introduces so many linguistic courses that it is necessary to review them many times. The big picture of what the objective of the course is only becomes clear (at least to me) during the last month of lectures. 

The readings are heavy, and sometime not all of them are relevant. I recommend re-reading the readings of the first weeks and also the ones for the last weeks. This is a luxury given the amount of load at the end of the semester, so be strategic about it.

The project is to make a lexical entry using a written corpus and the concepts from the course. The wikipedia entry on [Lexical functions](https://en.wikipedia.org/wiki/Lexical_function) is really helpful as a model of what is expected for some parts of the entry.

This course is shared with the *Science du Langage* students.

## Syntactic structures

**Where?**
- *Campus Lettres*

**Grading**
- 2 Exams

**Description**

Companion course to Lexicology, focusing on Dependency Syntax and Semantics. Several exercises are solved in each session.

**Topics**

- Parts of speech
- Syntactic dependencies, trees
- Semantic representations
- Lexical functions
- Syntactic coordination
- Projectivity
- Compound classes

**Comment**

The main idea of the course is to learn to create semantic graphs (a representation of a meaning to be communicated), and use rules to transform them into syntactic trees (the utterances that express the graphs' meaning).

The course uses several linguistic concepts, but the examples help to clarify them quickly.

This course is shared with the *Science du Langage* students.


## Semantic Web

**Where?**
- IDMC

**Grading**
- 1 Exam
- 1 Project

**Description**

An introduction to Knowledge graphs, how to create them, store them, query, and interconnect them. There are lectures and practical sessions.

**Topics**

- The Web, HTTP requests
- Resource Description Framework - RDF
- SPARQL
- RDF Schema, Description logics
- Ontologies
- Ontology reuse: Notable ontologies
- Applications using Semantic Web


**Comment**

The course is about how one can use the principles behind the World Wide Web to not only store and share data, but also to store and share knowledge. By knowledge, we mean a way of representing facts (like "Nancy is a City in France") in a way that's understandable for computers. Really interesting course.

Each lecture starts with a review of the previous one, and that makes the course feel very repetitive.

The practical sessions are very useful to ground the concepts from the lecture. Try to install all the tools (Blazergraph, Protege) before the practical sessions! 

The project consists on creating your own knowledge graph by mining and processing one from Wikidata or DBpedia. Then, you need to create a simple user interface to answer user questions, and an internal system that handels them and creates the appropriate query and presents the results to the user.

This course is shared with the Cognitive Science students.

# Project

There is one project going in parallel to the courses.

## Supervised Project

The supervised project is a **research project** which lasts for the first **year**. 

During the second semester you are supposed to implement and idea after you have done your litterature review during the first semester. You have to write a report of your idea and how you implemented it, and defend it in front of a jury.

We discovered that, even if we had reviewed the litterature decently well, we only knew about the methods in a theoretical way: we did not know how to write code to actually do what we wanted. This was a very common situation. In the case of my group, it took several months of trial and error with different ideas and software tools, in order to come up with a pipeline that we could make work.

For the defense of the project, you have to make slides and a poster.

Again, you should be having meetings with your supervisor every one or two weeks to report on your progress and ideas for the project.

Again, the workload of this project will depend a lot on your group dynamic and your supervisor. Some supervisors were very hands-off while others had a very specific roadmap. Some groups had everyone collaborate while some others had a single student doing all of the work. 

# Advise

The second semester was less heavy than the first one in terms of course hours because we were not using the *Alternance* system. However, in terms of project load it was more intense than the first one.


> Eventually, one has to balance whether all the courses and projects are worth it.


Additionally, there is very little feedback from the professors about your performance (for example, we never got grades for our tutorial assignments), so you only know how well you've been doing when you get the final grade.
