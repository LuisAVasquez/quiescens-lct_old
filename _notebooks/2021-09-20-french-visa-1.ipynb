# What is Campus France?


image.png

Every country has its own educational system. How can the French Government know if the requirements of one country match the requirements in France? This is what Campus France exists for.

[Campus France](https://www.campusfrance.org/en) is a service from the French Government dedicated to:
- handling the applications of foreing students to French Universities. Since I was coming from an Erasmus Mundus program, I did not have to go through the proccess of applying through Campus France.
- analyzing foreign academic qualifications (like university Diplomas) and decide on its equivalent in the French educational system.

After deciding on the qualifications of a candidate, they emit a certificate, which is later used by the French embassy for the Visa application.

Not all countries must go through the Campus France process when applying for a French Visa.
