After one year living in Nancy, one gets used to how the bureaucracy works. Still, there's bureaucracy to do before leaving.

# Leaving Residence Battelière

> You need to give **one month notice** at the reception before leaving CROUS residences. 

Ask about this at the reception, and they will give you a form and a link to book an appointment for a *état de lieu* ("state of the place"). Fill in the form and book the appointment at most one month before leaving the residence. The form is very simple: just state the day when you will leave. The *état de lieu* appointment is for checking everything is clean and in a good state in your room. On your last day you have to leave the keys at the reception. 

Additionally, **your CAF rights won't be affected**. If you leave in the middle of the month, you will get half the CAF for that month, if you leaving after the first third of the month, you'll get a third of the CAF for that month, and so on.

Note:
- The *état de lieu* can be done a couple of days before leaving the residence, but not that many.
- If your room is dirty, they will charge you for a cleaning service, which costs around 16€/hour. The amount of hours will depend on how dirty your place is.
- If there is something broken, they will charge you for replacing it. Remember that as soon as something breaks, you can ask for it getting repaired (check the post about living at Residence Batelière).


> In my case, it took around 3 months to get the deposit back!

# Closing Bank Accounts and Phone subscriptions

If you opened a bank account at a physical bank in Nancy (like BNP Paribas, Crédit Agricole, etc...), it may be convenient for you to close your account. 

Why? Because these banks are really old fashioned, so you are only able to do modifications to your account *in the same office where you opened it*. That means that, when the special rate for students is no longer valid and they start charging you, you would have to come back to Nancy to close the account.

If you opened a bank account at a digital bank, then you are mostly fine, as everything can be done on their app. Remember that when you move to a new country, if you want to keep the account, you will have to send them your new residence permit.


> The decision is up to you!

Note: 
- You have to tell the CROUS where to send your deposit money back. It can be a French bank, or a European one, as long as the account has an IBAN. Ask about this at the reception of the CROUS residence.



Phone subscriptions are way more flexible. However, if you have a contract for a plan, they will need your address, so, if you move out of France, it will get complicated to keep the plan. Prepaid SIM cards are fine in this case.

# Moving out

Of course, the process of leaving France overlaps with the process of moving to your Year 2 University. Eventually, you will have to move your stuff to your next destination, so, here I summarize the most effective ways of doing this depending on where you are going, if you depart from Nancy.

## Moving from Nancy to Donostia (Spain)

This was my case. After researching a lot of possibilities, I decided to book trains and take advantage of the fact that Donostia is very close to the French border. My route was the following:

- From Nancy to Paris *Gare de l'Est* station by train. 1.5 hours on the fast train (or 4 hours on the regional train).
- From Paris *Gare de l'Est* station to Paris *Montparnasse* station using the Metro line 4 (the purple one). 1 to 2 hours.
- From Paris *Montparnasse* to [Hendaye](https://www.google.com/maps/place/Hendaye,+France/) on the fast train. 4 hours. 
   - Hendaye is a town right at the Spanish-French border.
   - (It is amazing that you can cross France on fast train in 4 hours!)
- From Hendaye to Donostia via tram. Less than one hour.

That's how I was able to move in one day and in a single trip, here's some details about the journey: 

- For these train tickets, I used [discount cards](https://luisavasquez.github.io/quiescens-lct/nancy/going%20to%20france/2021/10/26/transport-nancy.html#Discount-Cards) and booked two months in advance to get amazing prices (I moved for less than 100€).
- Trains don't have luggage limits, and you can look for trains that allow bikes.
- The most uncomfortable part of the journey is taking the Paris Metro.
  - There is an entry to the metro station inside *Gare de l'Est* but it is always crowded. It is easier to go outside and find another entrance.     
  - It takes a while to find an elevator, and the station is not friendly for carrying luggage.
  - The rumors about the Paris metro are true: it smells really bad
- The fast trains have a cart that sells snacks, but they're very pricey.
- Just after heading out from the Hendaye train station, you will see the tram station to Donostia. 
- The tram from Hendaye to Donostia costs around 6 €.
- The border police may check your documents before boarding the tram.
- I left Nancy around 8 AM and was in Donostia around 5 PM. 


Note: 
- If you prefer to move by bus, then I strongly recommend that you avoid transferring in Paris. Why? Because you will have to transfer at Paris-Bercy bus station, which is a very sketchy and unsafe place. 
  - For bus, I would recommend the following route: Nancy > South of France (Lyon, Montpellier, or Marseille) > Barcelona > Donostia.

## Moving from Nancy to Trento (Italy)

> I have heard from other students that for this journey one has to transfer in Milan.

You can book with Flixbus from Nancy to Trento:
- From Nancy to Milan. 8 hours. 
    - Better to do it overnight.
- From Milan to Trento/Roveretto. 3 hours.

Several buses in this route have bike racks. 

Note:
- Taking the train in Milan to Trento/Roveretto is more expensive and not much faster.
- The bus station in Milan is called Lampugano
- The Lampugano station is very active, so it is easy to miss your bus stop.

# Declaring taxes (Optional)

> Declaring taxes makes sense if you want to stay or come back to France

The main objective is to obtain a tax number in France. Most French pay their taxes online, but, since you're new in France, you will have to make your first tax declaration in person. Luckily, the process is really easy. First you have to go to the *Hôtel d'Impôts* (Tax office) in Nancy:

<iframe src="https://www.google.com/maps/embed?pb=!1m18!1m12!1m3!1d2633.517237837606!2d6.184500215668631!3d48.695593979271884!2m3!1f0!2f0!3f0!3m2!1i1024!2i768!4f13.1!3m3!1m2!1s0x47949814bf1c6123%3A0x1aa3495e352232d6!2sCentre%20des%20Finances%20Publiques%20de%20Nancy!5e0!3m2!1sen!2sit!4v1672041556255!5m2!1sen!2sit" width="400" height="300" style="border:0;" allowfullscreen="" loading="lazy" referrerpolicy="no-referrer-when-downgrade"></iframe>

Usually you would have to book an appointment, but you can just show up, be sure to go before noon though! Try to go with a native French speaker.

You have to take with you:
- your passport and a copy
- your visa VLS-TS and a copy
- your *attestation de residènce* (residence certificate), which you can get at the reception of the CROUS residence

In my case, after waiting in line for a while, I got to the clerk and said I wanted to declare taxes. I explained my situation:
- I moved to France less than one year ago
- I had not worked in France
- My main source of money was an Erasmus scholarship
- and other simple details.

Then, the clerk brought a form and helped me fill it. Luckily, student scholarships are not taxes, so I did not have to pay any taxes at all (maybe that's what made the process so simple, though).

The whole thing took less than half an hour. I would receive my tax number in ten weeks via physical mail at my address.

> The deadline for declaring your taxes in person is usually at the end of May!

# Other

## Still a student of the University of Lorraine

As part of the Erasmus Mundus program, when you move to your second year university, you will technically still be a student of the University of Lorraine. This means that, if you happen to be in France again during that year, you can enjoy the perks of being a French student, like discounts at museums, or some public transport.

- You will receive an email to do the [academic registration](https://luisavasquez.github.io/quiescens-lct/french%20bureaucracy/nancy/going%20to%20france/lct%20year%201/2021/11/02/registering_university_lorraine.html#Academic-registration) for the second year. This just takes a couple of minutes.

- After that, you can download your [proof of enrollment](https://luisavasquez.github.io/quiescens-lct/nancy/lct%20year%201/2021/11/05/uni_lorraine_platforms.html#Dossier-%C3%A9tudiant:-Grades-and-Proof-of-enrollment) for the second year.
