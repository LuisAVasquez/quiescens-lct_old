# General Information

image.png

## What's *CAF*?

The [*Caisse d'allocations familiales*](https://www.caf.fr/) (*CAF*) is the branck of the French social security system dedicated to assisting families. They also deal with housing subsidies for students.

## Who can get *CAF*?

Anyone with student status can apply for the *CAF* student housing subsidy. 

That means that you can get *CAF* even if you are not being housed by the *CROUS* (you will need more documents though).

## How much does *CAF* pay?

From what I understand, they calculate how much they will cover based on your income and accomodation expenses. In my case, they cover around 40% of the rent for my place.

## When should I apply to *CAF*?

As soon as you can, the process is *really*,  **really**, **REALLY** slow. I'd recommend doing it as soon as you get an [European Bank account](https://luisavasquez.github.io/quiescens-lct/french%20bureaucracy/nancy/going%20to%20france/2021/11/14/bank_account.html).


## How long does the online application last?

Prepare to be filling out the online forms for a whole afternoon.

## How long till they give me an answer?

Depends on how early you do it. I did it really early, at the end of September, and got an acceptance answer around the end of December. 

However, other friends who applied later had to wait six, eight or ten months.



## How long does the subsidy last?

If you get *CAF*, you're covered for an academic year. Students who study in France for longer periods need to apply again. I don't know if this is a different process.

## Is it worth to get *CAF*? What if I apply too late?

You'll notice the bureaucracy feels infinite, but it pays off. 

When I applied, the *CAF* subsidy was retroactive, meaning that you got it even for the months of the academic year before doing the application. Because of this, I paid **zero** rent for three months.

An if by the end of the academic year there are some funds left from your *CAF*, they will reimbourse you!




# Applying



image.png

The specific bereaucracy to be done at the *CAF* is called the *Laide personnalise au logement APL* ("personalised support for accomodation").

## Guides

Remember that this blog is just my personal experience, to give you an idea of what to more or less expect. For more detailed and exact information, there are [guides available](https://drive.google.com/drive/folders/1bn_6koGcEaQgARbp9U9s5ru42ALkDsMB?usp=sharing) (and in many languages!). However, these guides are for the old online platform, there may be some updated version in the *CAF* website.

## Necessary documents

I needed to present the following documents:

- Passport
- Student Visa + Visa Validation
- Birth certificate (I presented it with an official translation, but I don't know if this is absolutely necessary)
- *Attestation de loyer CROUS* (Certificate of accomodation at CROUS). This was given to me at key pick up. If you lose it you can get it again at the residence reception or download it online.

Additionally, it will be necessary to have
- an email address
- a [French phone number](https://luisavasquez.github.io/quiescens-lct/french%20bureaucracy/nancy/going%20to%20france/2021/10/31/telephone-france.html)

## Starting with a simulation, and creating an account

In the [*CAF* website](https://www.caf.fr/), the path for starting the application starts with doing a simulation.

The way to get there is to search for
-*Logement* (accomodation), then
- *tudiant* (student), then
- *je ne suis pas allocataire* (I am not a beneficiary), then
- *faire une simulation* (do a simulation)

Then you will be directed to a short simulation, where you will be asked things like

- postal code of your place
- type of residence (is it a CROUS residence?)
- type of room
- rent price
- are you living alone or in couple?
- are there any kids living with you?
- what has been your income?
- do you have a scholarship?

Then you'll be given an estimate of the aid you will get, and a button to do the request (*faire la demande*).

Then, you will be directed to a log in website. If you already have your [*Carte Vitale*](https://luisavasquez.github.io/quiescens-lct/french%20bureaucracy/nancy/going%20to%20france/2021/11/24/french_healtcare.html) (which is almost impossible at such an early stage of your time in France) you can use its number. Otherwise, like most international students, you will have to look for the *Premire connexion? - Crer Mon Compte* (first connection? - create my account) button at the bottom of the website.



Screenshot from 2022-05-30 17-04-36.png

Then, check there for the option *Vous n'tes pas allocataire* (you are not a beneficiary), where you will be directed to the website to *faire une demande* (do a request).

Then, you will be directed to a website called *Faire une demande de prestation* (do a benefits request), in the option *logement* (accomodation), you will see the button *faire la demande* (do the request).


Screenshot from 2022-05-30 17-03-18.png


Then, you will be directed to a website to create an account. Here you will have to
- fill in your personal data, including email and phone number
- confirm the account creation with a code sent you by SMS

## My application

My application had many sections, I will go through all the questions I had to answer.


#### Section 1: General information about the accomodation

- ZIP code
- are you living alone or in a couple?
- are you a student (even in a sabatical year)?. That is, do you have student status?
- are you living with any kids?
- do you have any roommates?
- date of entry into the residence
- are you fiscally dependent on your parents? Do you appear in their (French?) tax declaration?)
- are you asking for a subsidy for your main place of residence?
- do you pay rent for the place?
- is the place a property of your family?

#### Section 2: Personal information
- parents names
- date of entry to France (check the stamp on your Passport!)
- Civil status married, single, etc.?
- Professional status? - Student
- Do you have a "social scholarship"? (Erasmus or research scholarships **do not** count as social scholarships).
- Address as in the rental contract
   - The website has an auto-complete feature for the address. In my case, this was not working. I used the option to enter my address manually. 

#### Section 3: Characteristics of the place
Here you will be asked for a bunch of information from the *attestation de loyer CROUS*.

#### Section 4: Recapitulation
Check all the information is accurate.

#### Section 5: Banking details
Put your IBAN.




#### Confirmation of request reception

Finally, you will get an email with a confirmation that your request has been taken into consideration.



## Wait

Your application status will show "pending" for a long time, this is normal. However, if you suspect something did not go right, you can go talk to them.

Something really convenient is to go with a French speaker friend.

---

In my case, there had been a major redesign of the online platform while my application was still pending, so I was worried that it would have gotten lost or something like it. 

I went with a French speaker friend, taking my documents with me, and asked for help. 

I got lucky that the *CAF* official that took me was helpful, they assured me that my application was on track, then asked for my birth certificate and  my passport and added them to my file. Since I already had gotten my *Carte Vitale* by that time, I also asked them to register my social security number in my file.

# Acceptance

Many months after my application, I got a letter from the *CAF* saying I had become a beneficiary (*allocataire*)! This included my *allocataire* number.

After this, the following month I got the *CAF* subsidy for my rent.

image.png

# Recommendations

- Again, start the application as soon as you can
- If you have to go to the *CAF* office, go with someone that speaks French
- Link your *CAF* account to [FranceConnect](https://luisavasquez.github.io/quiescens-lct/french%20bureaucracy/nancy/going%20to%20france/2021/11/24/french_healtcare.html#Using-FranceConnect) to manage it more easily.
